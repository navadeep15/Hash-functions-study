# Hash Functions Comparative Study

This notebook renders plots for the report using CSVs generated by the CLI.


```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

bench = pd.read_csv(Path('results/bench.csv'))
bench.head()
```

```python
# Throughput chart
pivot = bench.pivot(index='input_size_bytes', columns='algorithm', values='mb_per_s')
pivot.plot(figsize=(8,4), title='Throughput (MB/s) vs Input Size')
plt.xscale('log'); plt.show()
```
